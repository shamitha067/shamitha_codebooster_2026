```python
import pandas as pd
```

```python
df = pd.read_csv('/content/student_performance.csv')
display(df.head())
```

```python
total_students = df.shape[0]
print(f'Total number of students: {total_students}')
```

```python
total_columns = df.shape[1]
print(f"total no of columns:{total_columns}")
```

```python
print("Total number of departments:", df['department'].nunique())
```

```python
department_student_count = df['department'].value_counts()
print("Department-wise student count:\n", department_student_count)
```

```python
print(df.head())
```

```python
print("Highest Math Score:", df['math_score'].max())
print("Lowest Math Score:", df['math_score'].min())
print("Average Math Score:", df['math_score'].mean())
print("Average Programming Score:", df['programming_score'].mean())
print("Average Attendance:", df['attendance_percentage'].mean())
```

```python
print("\n[4] GENDER-WISE AVERAGE SCORE")
print("-"*40)
gender_scores_analysis = df.groupby('gender')[['math_score', 'science_score', 'english_score', 'programming_score']].mean().round(2)
print(gender_scores_analysis)
```

```python
print("\n[5] TOP 5 PERFORMANCE")
print("-" * 40)

df['total_score'] = df[['math_score', 'science_score', 'english_score', 'programming_score']].sum(axis=1)
top5 = df.sort_values(by='total_score', ascending=False).head()

for index, row in top5.iterrows():
    print(row['name'], "-", row['total_score'])
```

```python
print("Students of Lower Attendance")
print("="*50)
low_at = df[df['attendance_percentage']<75][['name','department','attendance_percentage']]
if len(low_at)>0:
  print(low_at.to_string(index=False))
else:
  print("All the Students are Greater than 75 %")
```
