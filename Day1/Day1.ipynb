```python
print("Department in this dataset")
print(df['department'].unique())
print("\nhow many students per department")
print(df['department'].value_counts())
```

```python
cs_students = df[df['department']== 'Computer Science']
print(f"No of Computer Science:{len(cs_students)}")
print("\nComputer Science students:")
print(cs_students[['name','math_score','programming_score','attendance_percentage']])
```

```python
high_attendance_students = df[df['attendance_percentage'] > 90]
high_attendance_students_cs = high_attendance_students[high_attendance_students['department'] == 'Computer Science']
print(f"Number of Computer Science students with attendance above 90%: {len(high_attendance_students_cs)}")
print("\nComputer Science students with high attendance:")
print(high_attendance_students_cs[['name', 'department', 'attendance_percentage']])
```

```python
top_performing_all_subjects = df[
    (df['attendance_percentage'] > 90) &
    (df['math_score'] > 70) &
    (df['science_score'] > 70) &
    (df['english_score'] > 70) &
    (df['programming_score'] > 70)
]

print(f"No of students with attendance above 90% and all subject scores above 70%: {len(top_performing_all_subjects)}")
print("\nStudents with high attendance and top performance in all subjects:")
print(top_performing_all_subjects[['name', 'department', 'attendance_percentage', 'math_score', 'science_score', 'english_score', 'programming_score']])
```

```python
avg_math_score_per_department = df.groupby('department')['math_score'].mean()
print("Average Math Score per Department:")
print(avg_math_score_per_department)
```

```python
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']
dept_avg = df.groupby('department')['total_score'].mean()
print("average total score:")
print(dept_avg)
```

```python
dept_avg = df.groupby('department')['total_score'].mean()
print("Average Total Score per Department (Descending Order):")
print(dept_avg.sort_values(ascending=False))
```
